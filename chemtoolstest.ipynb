```python
from chemtools import chemdict
import numpy as np
from scipy.linalg import null_space
import sympy

reactants = ["C2H6", "O2"]
products = ["CO2", "H2O"]

react_dict_list = [chemdict(r) for r in reactants]
prod_dict_list = [chemdict(p) for p in products]

elements = {key for d in (react_dict_list + prod_dict_list) for key in d}
elements_dict = {el: i for i, el in enumerate(elements)}
print(f"elements_dict: {elements_dict}")

react_vectors = []
prod_vectors = []

for react in react_dict_list:
    vector = np.zeros(len(elements_dict))
    for k, v in react.items():
        vector[elements_dict[k]] = v
    react_vectors.append(vector)

for prod in prod_dict_list:
    vector = np.zeros(len(elements_dict))
    for k, v in prod.items():
        v *= -1
        vector[elements_dict[k]] = v
    prod_vectors.append(vector)

matrix = np.array(react_vectors + prod_vectors, dtype= "i")
matrix = matrix.T
ns_list = sympy.Matrix(matrix).nullspace()[0]
ns_list = ns_list * (1 / ns_list[0])
print(list(ns_list))
```

```python

if ns_list:
    ns = ns_list[0]
    print(f"Nullspace: {ns_list}")
    scalar, v_init = ns.as_content_primitive()
    print(f"Scalar {scalar}\nV: {v_init}")
    print("")
    answer = np.array(v_init).astype(int).flatten()
    print(answer.shape)

print(f"Answer: {answer}")

#factor = 1 / ns_list[0]
#print(ns_list * factor)
```

```python
from chemtools import molarweight

class Compound:
    def __init__(self, formula: str) -> None:
        self.formula = formula
        self.weight = molarweight(formula)

class Mixture:
    def __init__(self, volume: float, solvent:str) -> None:
        self.volume = volume
        self.solvent = solvent
        self.contents = dict()

    def add(self, compound: Compound, concentration: float) -> None:
        if type(compound) == Compound and isinstance(concentration, (int, float)):
            self.contents[compound] = concentration
        else:
            raise ValueError("Incorrect inputtype")
```
